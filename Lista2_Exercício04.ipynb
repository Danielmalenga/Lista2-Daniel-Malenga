```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

N=100000
x=np.random.uniform(0,1,N)
integral=sum(((1-(x**2))**1.5))/N



#Defini√ß√£o da Fun√ß√£o
def f(x):
  return (((1-(x**2))**1.5))

# Cria valores do eixo X
x = np.linspace(0, 1, 100)
print(f'Integral Monte Carlos = {integral:.4f}')

# Plota a fun√ß√£o F(X)
plt.plot(x, f(x))

plt.fill_between(x, f(x), color='lightblue')

plt.text(0.2, 0.4, f'Area = {integral:.3f}', fontsize=14)


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(' INTEGRA√á√ÉO -MONTE CARLOS')

plt.show()

#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
ygx = y**0.4
#calcular a integral
integral = np.sum(((((1-(ygx**2))**1.5)))/((5/2)*ygx**1.5)) / N
print(f'Integral Por Import√¢ncia = {integral:.4f}')
#plotar o grafico
x = np.linspace(0, 1, N)
curve_y = (((1-(x**2))**1.5))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='green')
plt.text(0.5, 0.8, f'Area = {integral:.4f}', fontsize=14, color='red')
plt.xlim(-0.01, 1.1)
plt.ylim(-0.01, 1.1)
plt.title(' INTEGRA√á√ÉO - POR IMPORT√ÇNCIA')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
