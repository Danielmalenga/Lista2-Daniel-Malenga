```python
import numpy as np
import matplotlib.pyplot as plt

N=10000
y=np.random.uniform(0,1,N)
x = 1/y - 1
integral= sum((x*((1 + x**2)**(-2)))/y**2)/N
print(f'Integra√ß√£o Por Monte Carlos = {integral:.3f}')
# Define the function
def f(x):
  return x*((1 + x**2)**(-2))

# Create the x-axis values
x = np.linspace(0, 30, 500)


# Plot the function
plt.plot(x, f(x))

# Fill the area under the curve
plt.fill_between(x, f(x), color='lightblue')

# Add the label with the area value
plt.text(5, 0.3,f'Area = {integral:.3f}', fontsize=14)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integra√ß√£o Por Monte Carlos')

# Show the plot
plt.show()

y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
A = 3/8
ygx = (((3*y)/A))**(1/3)
#calcular a integral
integral = np.sum((ygx*((1 + ygx**2)**(-2)))/(A*ygx**2)) / N
print(f'Integral Por Import√¢ncia = {integral:.4f}')

#plotar o grafico
x = np.linspace(0, 30, 500)
curve_y = x*((1 + x**2)**(-2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(5, 0.3,f'Area = {integral:.3f}', fontsize=14)
plt.title(' INTEGRA√á√ÉO - POR IMPORT√ÇNCIA')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
