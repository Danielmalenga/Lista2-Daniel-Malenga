```python
import numpy as np
import matplotlib.pyplot as plt

N=1000000
y=np.random.uniform(0,1,N)
x = 4*y - 2
integral=4*sum(np.exp((x+x**2)))/N

print(f'Integra√ß√£o Por Monte Carlos = {integral:.3f}')
# Define the function
def f(x):
  return np.exp((x + x**2))

# Create the x-axis values
x = np.linspace(-2, 2, 50)


# Plot the function
plt.plot(x, f(x))

# Fill the area under the curve
plt.fill_between(x, f(x), color='lightblue')

# Add the label with the area value
plt.text(-1, 250,f'Area = {integral:.3f}', fontsize=14)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(' INTEGRA√á√ÉO -MONTE CARLOS')
# Show the plot
plt.show()

###Import√¢ncia

y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
A = 1/(np.exp(3) - np.exp(-1))
ygx = np.log(y/A + (np.exp(-1))) -1

#calcular a integral
integral = np.sum((np.exp(ygx + ygx**2)) /(A*np.exp(ygx+1))) / N
print(f'Integral Por Import√¢ncia = {integral:.3f}')
#plotar o grafico
x = np.linspace(-2, 2, N)
curve_y = np.exp((x + x**2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(-1, 200, f'Area = {integral:.4f}', fontsize=14, color='black')
plt.title(' INTEGRA√á√ÉO - POR IMPORT√ÇNCIA')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
